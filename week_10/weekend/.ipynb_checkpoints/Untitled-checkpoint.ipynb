*Question 1*

Write a function that takes in a list and returns a two-element list containing the first and last elements of the inputted list. For example, for input [3, 6, 2,  19], the function should return [3, 19]

```python
def two_element_list(input_list):
    
    return list([input_list[0], input_list[-1]])
    
    
```

```python
my_list = [3, 6, 9, 12, 19]

two_element_list(my_list)
```

_Question 2_

_Write a function that takes in a sentence and returns a dictionary containing a count of letters and digits. For example, for input "hello world! 123", the function output should be {'digits': 3, 'letters': 10}_


```python
"Hello world! 123"
```

```python
def sentence_to_dict(string):
        
        
        numbers = sum(i.isdigit() for i in string)
        words   = sum(i.isalpha() for i in string)
        spaces  = sum(i.isspace() for i in string)
        others  = len(string) - numbers - words - spaces

        return {"numbers": numbers,
                "letters": words, 
                "spaces":spaces,
                "others": others
               }
```

```python
sentence_to_dict("Hello world! 123")
```

```python
sentence_to_dict("I want to ride my bicycle, I want to ride my bike")
```

_Question 3_

Write a class to model a bank account in a simple way. The class should have a balance attribute that starts at a value zero when the account is created, along with methods: 

- get_balance() - prints balance to screen
- deposit(amount) - should deposit any positive amount and print new balance to screen, or print an error if amount is negative or zero.
- withdraw(amount) - should work only if there are sufficient funds in the account, withdraw any positive amount and print new balance to screen, otherwise it should print an error message.

Create your own bank account object from your class and test that all of your methods work as expected, including testing for expected errors.

[Hint - remember to pass self into all methods in the class]

```python
class Bank_Account:
    
    def __init__(self, balance):
        self.balance = balance 
        
    def get_balance(self):
        print("£" + str(self.balance))
        
    def deposit(self, amount): 
        
        if amount > 0: 
            self.balance += amount
            print("New balance: £" + str(self.balance))
            
        else:
            print("Error: please select an amount greater than 0.")
            
    def withdraw(self, amount):
        if amount > 0 and amount < self.balance: 
            self.balance -= amount 
            print("New balance: £" + str(self.balance))
        
        else:
            print("Your account has insufficient funds for this transaction.")
            
            
    

    
```

```python
my_account = Bank_Account(1200)
```

```python
my_account.get_balance()
```

```python
my_account.deposit(100)
```

```python
my_account.withdraw(10)
```

```python
my_account.withdraw(1300)
```
